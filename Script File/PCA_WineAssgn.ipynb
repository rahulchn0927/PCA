```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale 
```

```python
wine=pd.read_csv(r'C:\Users\RAHUL CHAUHAN\Downloads\wine.csv')
wine.head()
```

```python
wine.describe()
```

```python
wine.shape
```

```python
wine_data=wine.iloc[:,1:]
wine_data.head()
```

```python
WINE=wine_data.values
WINE
```

```python
win_norm=scale(WINE)
win_norm
```

# PCA

```python
pca=PCA()
pca_values = pca.fit_transform(win_norm)
```

```python
pca_values
```

```python
pca.components_
```

```python
var=pca.explained_variance_ratio_
var
```

```python
var1=np.cumsum(np.round(var,decimals=4)*100)
var1
```

```python
# Variance plot for PCA components obtained 
plt.plot(var1,color='Blue')
```

```python
# Final Dataframe
final_df=pd.concat([wine['Type'],pd.DataFrame(pca_values[:,0:3],columns=['PC1','PC2','PC3'])],axis=1)
final_df
```

```python
# Visualization of PCAs
fig=plt.figure(figsize=(16,12))
sns.scatterplot(data=final_df)
```

# Checking clustering Algorithms

# Heirarchical clustering

```python
# import hierarchical clustering libraries
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
#check dendrogram
plt.figure(figsize=(20,8))
dendrogram=sch.dendrogram(sch.linkage(pca_values,'complete'))
```

```python
# Create Clusters (y)
hclusters=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='complete')
hclusters
```

```python
y=pd.DataFrame(hclusters.fit_predict(pca_values),columns=['Clusters'])
y['Clusters'].value_counts()
```

```python
wine2=wine.copy()
wine2['Clusters']=hclusters.labels_
wine2
```

# KMeans Clustering

```python
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer
```

```python
Sum_of_squared_distances =[]
K = range(1,6)
for k in K:
    km =KMeans(n_clusters =k)
    km =km.fit(pca_values)
    Sum_of_squared_distances.append(km.inertia_)
    
plt.figure(figsize=(10,7))

plt.plot(K, Sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('SSE')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
#Elbow graph depicts optimum no. of clusters = 3
```

```python
# Cluster algorithm using K=3
clusters=KMeans(3,random_state=30).fit(pca_values)
clusters
```

```python
clusters.labels_
```

```python
wine3=wine.copy()
wine3['Clusters']=clusters.labels_
wine3
```
